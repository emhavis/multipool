```python
import requests
import pandas as pd

df1 = pd.read_csv('train_dataset_paper.csv')
df2 = pd.read_csv('test_dataset_paper.csv')

def run_curl_command(prompt):
    url = "http://108.137.109.168:11434/api/generate"

    data = {
        "model": "mistral",
        "prompt": prompt,
        "stream": False
    }

    # Sending a POST request
    response = requests.post(url, json=data)

    # Checking the response
    if response.status_code == 200:
        # Return the JSON response
        return response.json()
    else:
        print(f"Error! Status code: {response.status_code}, Response: {response.text}")
        # Return None or raise an exception, depending on your preference
        return None
```

```python
run_curl_command("explain this {x}"
                 .format(x=df1.head()))
```

```python
prompt = """
I have a test data in this "{y}"
Please respond in a valid JSON format by the train data in "{x}", group them by the following example:
by category and show the confidence level of each categorization example format:

  "title": "Reconstructing Subject-Specific Effect Maps",
  "category" : "Computer Science"
  "confidence level": 1.0
  "reason of confidence level": "The paper introduces a novel computational method, the "Reconstructing Subject-Specific Effect Maps" (RSM), aimed at improving subject-specific inference in neuroimaging data analysis. besides, The paper employs computational methods to process and analyze large-scale datasets, including synthetically generated data and data from the Alzheimer's Disease Neuroimaging Initiative (ADNI) database."
""".format(x=df1, y=df2)
```

```python
run_curl_command(prompt)
```
