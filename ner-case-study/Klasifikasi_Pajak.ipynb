```python
import requests
import pandas as pd
import re
```

```python
df1 = pd.read_csv('Sawit_Dataset.csv')
df2 = pd.read_csv('clean_data.csv')
def run_curl_command(prompt):
    url = "http://108.137.109.168:11434/api/generate"

    data = {
        "model": "mistral",
        "prompt": prompt,
        "stream": False
    }

    # Sending a POST request
    response = requests.post(url, json=data)

    # Checking the response
    if response.status_code == 200:
        # Return the JSON response
        return response.json()
    else:
        print(f"Error! Status code: {response.status_code}, Response: {response.text}")
        # Return None or raise an exception, depending on your preference
        return None
```

```python
def delete_all_characters_in_dataframe(df1):
    # Iterasi melalui setiap kolom dalam DataFrame
    for col in df1.columns:
        # Hanya memproses kolom yang memiliki tipe data string
        if df1[col].dtype == 'object':
            # Menghapus semua jenis karakter menggunakan ekspresi reguler
            df1[col] = df1[col].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)))
    return df1

dataset_clean=delete_all_characters_in_dataframe(df1)

print(dataset_clean)

```
